```python
import os, sys
from random import shuffle
# append the parent directory to the system path using current working directory instead of __file__
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
from pickle import load

# unpicke alpha-poker/temp/checkpoint_0.pth.tar.examples  
trainExamplesHistory = load(open('temp/checkpoint_0.pth.tar.examples', 'rb'))
print(trainExamplesHistory)
trainExamples = []
for e in trainExamplesHistory:
    # e = [e.pop() for _ in range(len(e))]
    # print(e)
    trainExamples.extend(e)
print(trainExamples)
print(len(trainExamples))
```

```python
community_cards = set([])
hole_cards = set([])
round_counts = {}
num_without_community_cards = 0

# TODO: for some reason, it seems like its stopping before getting to the community card portion. Maybe because it is folding too much and too early??
for state, pi, v in trainExamples:
    uuid = list(state.player_states.keys())[0]
    community_cards.update(state.player_states[uuid].community_cards)
    if len(state.player_states[uuid].community_cards) == 0:
        num_without_community_cards += 1
    hole_cards.update(state.player_states[uuid].hole_cards)
    round_count = state.player_states[uuid]["round_count"]
    if round_count not in round_counts:
        round_counts[round_count] = 0
    round_counts[round_count] += 1

print("Community cards: ", community_cards)
print("Hole cards: ", len(hole_cards))
print("Round counts: ", round_counts)
print("Number of examples without community cards: ", num_without_community_cards)
```
